**3. Генерация датасета и рассчет доверительного интервала**



```python
import numpy as np
import scipy.stats as stats

# Генерируем 500 случайных значений, представляющих игроков
np.random.seed(0)  # Фиксируем seed для воспроизводимости
total_players = 500
payment_players = np.random.choice([0, 1], size=total_players, p=[0.9, 0.1])

# Вычисляем долю плательщиков в сгенерированном датасете
payment_ratio = np.mean(payment_players)
print("Доля плательщиков в контрольной группе:", payment_ratio)

# Вычисляем доверительный интервал для доли плательщиков
conf_int = stats.norm.interval(0.95, loc=payment_ratio, scale= np.sqrt(payment_ratio*(1-payment_ratio)/total_players))
print("95% доверительный интервал для доли плательщиков:", conf_int)
```

Чтобы решить эту задачи методом bayesian нам потребуется выбрать априорное распределение для вероятности успеха и обновить его на основе имеющихся данных с использованием теоремы Байеса.

Предположим, что априорное распределение для вероятности успеха - бета-распределение, которое является сопряженным априорным распределением для биномиального распределения.

```python
import numpy as np
import scipy.stats as stats

# Генерируем 500 случайных значений, представляющих игроков
np.random.seed(0)  # Фиксируем seed для воспроизводимости
total_players = 500
payment_players = np.random.choice([0, 1], size=total_players, p=[0.9, 0.1])

# Вычисляем количество плательщиков
n_success = np.sum(payment_players)

# Вычисляем байесовскую оценку для вероятности успеха (используем бета-распределение в качестве априорного)
alpha_prior = 1
beta_prior = 1
alpha_post = alpha_prior + n_success
beta_post = beta_prior + total_players - n_success
posterior_dist = stats.beta(alpha_post, beta_post)

# Строим байесовский доверительный интервал
bayesian_conf_int = posterior_dist.interval(0.95)
print("95% байесовский доверительный интервал для доли плательщиков:", bayesian_conf_int)
```

Здесь мы используем бета-распределение как априорное распределение вероятности успеха, обновляем его по данным и вычисляем байесовский доверительный интервал для доли плательщиков. Этот подход позволяет учесть априорные знания и провести обновление вероятности успеха на основе имеющихся данных.

**4.Сгенерируйте несколько вариантов экспериментальной группы (хуже, лучше, без эффекта), рассчитайте силу эффекта, ДИ и ваши выводы для каждого варианта.**

Мы можем создать несколько вариантов экспериментальной группы, отражающих разные уровни эффективности обновления, и затем рассчитать силу эффекта для каждого варианта. Давайте определим три варианта:

1. "Хуже" - уровень плательщиков остается на уровне 10% или уменьшается.
2. "Лучше" - уровень платежей увеличивается до 11%.
3. "Без эффекта" - уровень плательщиков остается на уровне 10.5%.

Для каждого варианта мы рассчитаем силу эффекта и 95% доверительный интервал (ДИ) для разницы в долях плательщиков до и после обновления.


```python
import numpy as np
import scipy.stats as stats

# Зададим параметры
total_players = 500
baseline_payment_rate = 0.10  # 10% в контрольной группе
improvement = 0.01  # ожидаемое увеличение до 11% в экспериментальной группе
effect_sizes = []

# Генерируем данные для каждого варианта экспериментальной группы
for scenario in ['worse', 'better', 'no_effect']:
    if scenario == 'worse':
        # У худшего варианта ожидаемый процент уменьшается на 1%
        exp_payment_rate = baseline_payment_rate - 0.01
    elif scenario == 'better':
        # У лучшего варианта ожидаемый процент увеличивается на 1%
        exp_payment_rate = baseline_payment_rate + 0.01
    else:
        # У варианта без эффекта ожидаемый процент остается прежним
        exp_payment_rate = baseline_payment_rate

    # Генерируем данные для экспериментальной группы
    payment_players_exp = np.random.choice([0, 1], size=total_players, p=[1-exp_payment_rate, exp_payment_rate])

    # Вычисляем доли плательщиков
    obs_payment_rate = np.mean(payment_players_exp)

    # Вычисляем статистическую разницу в долях плательщиков
    effect_size = obs_payment_rate - baseline_payment_rate
    effect_sizes.append(effect_size)

    # Вычисляем стандартное отклонение и стандартную ошибку среднего для 95% доверительного интервала
    std_dev = np.sqrt((obs_payment_rate * (1 - obs_payment_rate) + baseline_payment_rate*(1 - baseline_payment_rate)) / total_players)
    margin_of_error = 1.96 * std_dev  # 95% ДИ

    # Вывод результатов
    print(f"{scenario.capitalize()} scenario:")
    print(f"Ожидаемый процент плательщиков: {exp_payment_rate*100:.2f}%")
    print(f"Фактический процент плательщиков: {obs_payment_rate*100:.2f}%")
    print(f"Сила эффекта: {effect_size*100:.2f}%")
    print(f"95% ДИ для разницы в долях плательщиков: [{(effect_size - margin_of_error)*100:.2f}%, {(effect_size + margin_of_error)*100:.2f}%]")
    print("\n")
```

   - Вывод: Полученная разница в долях плательщиков находится в пределах 95% доверительного интервала, что указывает на то, что изменение не обладает статистической значимостью. Мы не можем утверждать, что изменение привело к значимому ухудшению доли плательщиков.

Метрики Precision, Recall и F1-score: Эти метрики позволяют оценить баланс между полнотой (способностью модели обнаруживать положительные классы) и точностью (способностью модели давать правильные положительные предсказания). Precision вычисляет долю правильно предсказанных положительных классов относительно всех положительных предсказаний, Recall вычисляет долю правильно предсказанных положительных классов относительно всех фактически положительных классов, а F1-score является средним гармоническим между Precision и Recall. Эти метрики особенно полезны, если классы несбалансированы.

4. Кривая ROC (Receiver Operating Characteristic) и показатель AUC (Area Under the Curve): Кривая ROC визуализирует отношение между долей верных положительных предсказаний и долей ложных положительных предсказаний при изменении порога классификации. AUC показывает площадь под кривой ROC и является мерой качества модели. Чем выше AUC, тем лучше модель способна различать между классами.

Для данной задачи метрики Precision, Recall и F1-score могут быть более оптимальными. Причина в том, что эти метрики учитывают как правильные положительные предсказания, так и правильные отрицательные предсказания.

В данном случае, важно минимизировать ложные отрицательные предсказания, то есть ситуации, когда модель не предсказала, что игрок уйдет, а он все же ушел. Это позволит запустить альтернативное продолжение туториала на предшествующем шаге и, таким образом, избежать оттока игроков.
